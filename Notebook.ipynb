```python
class book:
    read_file = '/Users/w1203865668/Desktop/实训/day01/kaifangX.txt'
    write_file = '/Users/w1203865668/Desktop/text.txt'
    write = open(write_file,'w',encoding='gbk')
    with open(read_file,'r',encoding='gbk',errors='ignore') as f:
        for i in range(10000):
            try:
                email = f.readline().split(',')[9]
                write.write(email)
            except Exception as e:
                print('空白')
    write.close()
```

获取某网站的两个网页并保存到本地

```python
url = ['http://www.17k.com/book/2819588.html','http://www.17k.com/book/2793405.html']
see = '/Users/w1203865668/Desktop/text.html'
with open(see,'a',encoding='utf-8') as f:
    for i in range(2):
        import requests
        header = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.79 Safari/537.36'}
        b=requests.get(url[i],headers=header)
        f.write(b.text)
```
